```python
class GarbageDuctSystem:
    def __init__(self):
        self.active_floor = None
        self.duct_status = {'floor': None, 'status': 'closed'}
        self.uv_status = 'not_activated'
        self.fill_level = 0.0

    def use_duct(self, floor):
        if self.duct_status['status'] == 'closed':
            self.duct_status = {'floor': floor, 'status': 'open'}
            self.active_floor = floor
            print(f'Duct opened on floor {floor}')
        else:
            print(f'Duct is currently in use by floor {self.active_floor}')

    def close_duct(self, floor):
        if self.duct_status['floor'] == floor:
            self.duct_status = {'floor': None, 'status': 'closed'}
            self.active_floor = None
            print(f'Duct closed on floor {floor}')
        else:
            print(f'Error: Duct on floor {floor} is not open')

    def activate_uv(self):
        self.uv_status = 'activated'
        print('UV sanitization activated')

    def deactivate_uv(self):
        self.uv_status = 'not_activated'
        print('UV sanitization deactivated')

    def check_fill_level(self, level):
        self.fill_level = level
        if self.fill_level > 0.8:
            print('Alert: Garbage tank is almost full')
        else:
            print('Garbage tank fill level is normal')
```

```python
# Example usage
system = GarbageDuctSystem()
system.use_duct(5)
system.check_fill_level(0.85)
system.activate_uv()
system.close_duct(5)
system.deactivate_uv()
```
